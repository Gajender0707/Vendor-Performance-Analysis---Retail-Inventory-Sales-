### 5. Confidence Interval for the profit margin for the top performing and low perfoming vendor  

```python
data["UnsoldInventoryValue"]=(data["TotalSalesQuantity"]-data["TotalPurchaseQuantity"])*data["TotalPurchaseDollars"]
```

```python
top_peforming_vendor_threasold=data["TotalSalesDollars"].quantile(0.75)
```

```python
low_peforming_vendor_threasold=data["TotalSalesDollars"].quantile(0.25)
```

```python
top_perfoming_vendor=data[data["TotalSalesDollars"]>=top_peforming_vendor_threasold]["ProfitMargin"]
low_perfoming_vendor=data[data["TotalSalesDollars"]<low_peforming_vendor_threasold]["ProfitMargin"]
```

```python
from scipy import stats
def confidence_interval(data,confidence=0.95):
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample std
    n = len(data)
    # Confidence level
    confidence = 0.95
    z_score = stats.norm.ppf((1 + confidence) / 2)
    
    # Standard error
    se = sample_std / np.sqrt(n)
    
    # Confidence Interval
    ci_lower = sample_mean - z_score * se
    ci_upper = sample_mean + z_score * se
    
    print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
confidence_interval(top_perfoming_vendor)
confidence_interval(low_perfoming_vendor)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


# Function to calculate and visualize confidence interval
def confidence_interval_with_plot(data, vendor_name="Vendor", confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)

    z_score = stats.norm.ppf((1 + confidence) / 2)
    se = sample_std / np.sqrt(n)

    ci_lower = sample_mean - z_score * se
    ci_upper = sample_mean + z_score * se

    print(f"{vendor_name} - {int(confidence*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

    # Plot
    sns.set(style="whitegrid")
    plt.figure(figsize=(10, 4))
    sns.histplot(data, kde=True, color="skyblue", bins=8)

    plt.axvline(sample_mean, color='green', linestyle='--', label=f'Mean: {sample_mean:.2f}')
    plt.axvline(ci_lower, color='red', linestyle='--', label=f'CI Lower: {ci_lower:.2f}')
    plt.axvline(ci_upper, color='red', linestyle='--', label=f'CI Upper: {ci_upper:.2f}')
    plt.fill_betweenx([0, plt.gca().get_ylim()[1]], ci_lower, ci_upper, color='red', alpha=0.2)

    plt.title(f'{vendor_name} - {int(confidence*100)}% Confidence Interval')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Run for both vendors
confidence_interval_with_plot(top_perfoming_vendor, vendor_name="Top Performing Vendor")
confidence_interval_with_plot(low_perfoming_vendor, vendor_name="Low Performing Vendor")
```

1. The confidence interval for low-performing vendors (41.53% to 43.53%) is significantly higher than that of top-performing vendors (31.41% to 32.21%).
2. This suggests that vendors with lower sales tend to maintain higher profit margins, potentially due to premium pricing or lower operational costs.
3. For High-Performing Vendors: If they aim to improve profitability, they could explore selective price adjustments, cost optimization, or bundling strategies.
4. For Low-Performing Vendors: Despite higher margins, their low sales volume might indicate a need for better marketing, competitive pricing, or improved distribution strategies.
